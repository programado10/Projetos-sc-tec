## Importar CSV & Vizualização

```python
import csv

#---------------------------------------------------------------
# FUNÇÃO PARA LER ARQUIVOS CSV
#---------------------------------------------------------------

def ler_csv(nome_arquivo):
    dados = []

    with open(nome_arquivo, 'r', encoding='utf-8') as arquivo_csv:
        leitor_csv = csv.DictReader(arquivo_csv)

        for linha in leitor_csv:
            dados.append(linha)

    return dados


#---------------------------------------------------------------
# LENDO O DATASET DE PRODUTOS
#---------------------------------------------------------------

produtos = ler_csv('data information product.csv')





#---------------------------------------------------------------
# VERIFICANDO SE OS DADOS FORAM CARREGADOS
#---------------------------------------------------------------

print("Total de produtos:", len(produtos))





#---------------------------------------------------------------
# IMPRIMINDO AS PRIMEIRAS 7 LINHAS DOS PEDIDOS
#---------------------------------------------------------------

for linha in produtos[:7]:
    print(linha)



    
```

## UTILIZAÇÃO DO REGEX PARA PADRONIZAR A ESCRITAS NAS COLUNAS

```python
#---------------------------------------------------------------
#importar a biblioteca re para manipulação de strings
#---------------------------------------------------------------
import re

#---------------------------------------------------------------
#limpar a coluna "product_category_name" para padronizar os dados
#---------------------------------------------------------------

for linhas in produtos:


    #---------------------------------------------------------------
    #padronização de dados na coluna "product_category_name"
    #---------------------------------------------------------------
    
    linhas["product_category_name"] = linhas["product_category_name"].strip().lower()                         # Remove espaços em branco e converte para minúsculas
    linhas["product_category_name"] = re.sub(r"[^a-zA-ZÀ-ÿ0-9_ ]", "", linhas["product_category_name"])       # Remove caracteres especiais, exceto letras, números, espaços e sublinhados
    linhas["product_category_name"] = re.sub(r"\s+", "_", linhas["product_category_name"])                    # Substitui espaços por sublinhados
 
    #---------------------------------------------------------------
    #substituir valores vazios por "Sem_categoria"
    #---------------------------------------------------------------

    if linhas["product_category_name"] == "":
        linhas["product_category_name"] = "Sem_categoria"
    else:
        linhas["product_category_name"] = linhas["product_category_name"]

    #---------------------------------------------------------------
    #quantiade de produtos sem categoria
    #---------------------------------------------------------------

cont_sem_categoria = []
for linhas in produtos:
    if linhas["product_category_name"] == "Sem_categoria":
        cont_sem_categoria.append(linhas)

   #----------------------------------------------------------------
   #printar a quantidade de produtos sem categoria
   #----------------------------------------------------------------

print("Quantidade de produtos sem categoria:", len(cont_sem_categoria))


#----------------------------------------------------------------------------------------------------------------------------------------------------
#toda vez que for executar esse codigo, execute antes a celula de importação do arquivo csv, para garantir que os dados estejam atualizados e limpos.
#---------------------------------------------------------------------------------------------------------------------------------------------------





    
  



      



 
```

## CRIANDO FUNÇÃO PARA CALCULAR A MEDIA DAS COLUNAS DE PESO E DIMENSÃO COM VALORES NULOS

```python
#---------------------------------------------------------------
# FUNÇÃO PARA TRATAR COLUNAS COM VALORES VAZIOS USANDO A MÉDIA
#---------------------------------------------------------------

#-------------------------------------------------
#explicação sobre o motivo em que utilizei a media
#-------------------------------------------------
# Optei por substituir os valores vazios das dimensões físicas pela média da própria coluna.

# Essa escolha evita descartar produtos do dataset e impede o uso de valores irreais, como zero.

# A média representa um valor aproximado do padrão dos produtos cadastrados.

# Em um contexto logístico, como cálculo de frete, armazenamento ou escolha do veículo de transporte,

# conhecer dimensões médias ajuda a estimar se os produtos tendem a ser pequenos, médios ou grandes.


def tratar_coluna_com_media(produtos, nome_coluna):
    add = 0
    validos = 0

    # Primeiro: calcular a média
    
    for linha in produtos:
        valor = linha[nome_coluna].strip()

        if valor != "":
            add += float(valor)
            validos += 1

    media = add / validos

    # Segundo: substituir vazios pela média

    quantidade_corrigidos = 0

    for linha in produtos:
        valor = linha[nome_coluna].strip()

        if valor == "":
            linha[nome_coluna] = media
            quantidade_corrigidos += 1

# Imprimir informações sobre o tratamento

    print("Coluna tratada:", nome_coluna)
    print("Média usada:", media)
    print("Valores corrigidos:", quantidade_corrigidos)
    print("-" * 40)

```

## TRATANDO AS COLUNAS DE DIMENÇÕES

```python
#---------------------------------------------------------------
# tratar as colunas dimensionais com a média feita pela a função anteriomente
#---------------------------------------------------------------

#coloquei dentro das colunas dimensionais, as colunas de comprimento, altura e largura do produto.

colunas_dimensionais = ["product_length_cm", "product_height_cm", "product_width_cm"]

#executar a função para cada coluna dimensional

for coluna in colunas_dimensionais:
    
    tratar_coluna_com_media(produtos, coluna)

#antes de tratar as colunas dimensionais, execute a célula de importação do arquivo csv para garantir que os dados estejam atualizados e limpos.
```

## TRANTANDO A COLUNA DE PESO


```python
#---------------------------------------------------------------
#tratar a coluna de peso usando a média com a função criada anteriormente
#---------------------------------------------------------------

#troquei o nome product_weight_g para coluna peso, para ficar mais claro e facil de entender.

colunas_peso = ["product_weight_g"] 

#chamei a função tratar_coluna_com_media para cada coluna de peso, para substituir os valores vazios pela média calculada.

for coluna in colunas_peso:

 tratar_coluna_com_media(produtos, coluna)

#antes de tratar a coluna de peso, execute a célula de importação do arquivo csv para garantir que os dados estejam atualizados e limpos.
    
```

## LENDO O OUTRO ARQUIVO PARA OS PEDIDOS

```python
#---------------------------------------------------------------
# LEITURA DO DATASET DE PEDIDOS
#---------------------------------------------------------------

pedidos = ler_csv('data request product.csv')


#---------------------------------------------------------------
# VERIFICAÇÃO DA QUANTIDADE DE PEDIDOS CARREGADOS
#---------------------------------------------------------------

print("Total de pedidos:", len(pedidos))


#---------------------------------------------------------------
# EXIBIÇÃO DAS 7 PRIMEIRAS LINHAS DO DATASET
#---------------------------------------------------------------

print("\n--- Primeiras 7 linhas do dataset de pedidos ---")

for linha in pedidos[:7]:
    print(linha)
    print("-" * 40)


#---------------------------------------------------------------
# EXIBIÇÃO DOS NOMES DAS COLUNAS DO DATASET
#---------------------------------------------------------------

print("\n--- Colunas do dataset de pedidos ---")
print(pedidos[0].keys())
```

## ANALISE PEDIDOS 

```python
#---------------------------------------------------------------
# ANÁLISE DOS PEDIDOS SEM DATA DE ENTREGA
#---------------------------------------------------------------

total_pedidos_processados = 0
total_pedidos_cancelados = 0
total_pedidos_sem_data = 0
total_pedidos_sem_data_status_cancelado = 0
total_pedidos_sem_data_status_diferente = 0

# Lista para guardar alguns exemplos de pedidos sem data de entrega
# mas com status diferente de "canceled"
exemplos_datas_vazias_outro_status = []


for linha in pedidos:
    total_pedidos_processados += 1

    status = linha["order_status"].strip().lower()
    data_entrega_cliente = linha["order_delivered_customer_date"].strip()

    # Conta todos os pedidos cancelados
    if status == "canceled":
        total_pedidos_cancelados += 1

    # Verifica se a data de entrega está vazia
    if data_entrega_cliente == "":
        total_pedidos_sem_data += 1

        # Verifica se a data vazia pertence a um pedido cancelado
        if status == "canceled" and data_entrega_cliente == "":
            total_pedidos_sem_data_status_cancelado += 1

        # Caso a data esteja vazia, mas o status não seja cancelado
        else:
            total_pedidos_sem_data_status_diferente += 1

            # Guarda apenas os 5 primeiros exemplos para não poluir o relatório
            if len(exemplos_datas_vazias_outro_status) < 5:
                exemplos_datas_vazias_outro_status.append(linha)


#---------------------------------------------------------------
# RELATÓRIO DA ANÁLISE DOS PEDIDOS
#---------------------------------------------------------------

print("------ RELATÓRIO DE PEDIDOS ------")
print("Total de pedidos processados:", total_pedidos_processados)
print("Total de pedidos cancelados:", total_pedidos_cancelados)
print("Total de pedidos sem data de entrega:", total_pedidos_sem_data)
print("Total de pedidos sem data de entrega e status 'canceled':", total_pedidos_sem_data_status_cancelado)
print("Total de pedidos sem data de entrega e status diferente de 'canceled':", total_pedidos_sem_data_status_diferente)
print("----------------------------------")


#---------------------------------------------------------------
# EXEMPLOS DE PEDIDOS SEM DATA E COM STATUS DIFERENTE DE CANCELED
#---------------------------------------------------------------

print("\nExemplos de pedidos sem data de entrega e status diferente de 'canceled':")

for exemplo in exemplos_datas_vazias_outro_status:
    print("ID do pedido:", exemplo["order_id"])
    print("Status:", exemplo["order_status"])
    print("Data de entrega:", exemplo["order_delivered_customer_date"])
    print("-" * 40)
```

## CONVERSÃO DATETIME

```python
#---------------------------------------------------------------
# IMPORTAÇÃO DA BIBLIOTECA DATETIME
#---------------------------------------------------------------

from datetime import datetime


#---------------------------------------------------------------
# FORMATAÇÃO DA COLUNA order_approved_at
#---------------------------------------------------------------

datas_de_aprovacao_vazias = 0
datas_aprovacao_formatadas = 0

for linha in pedidos:
    data_aprovacao = linha["order_approved_at"].strip()

    # Verifica se a data de aprovação está vazia
    if data_aprovacao == "":
        datas_de_aprovacao_vazias += 1

    # Caso tenha data, converte para o formato brasileiro
    else:
        data_convertida = datetime.strptime(data_aprovacao, "%Y-%m-%d %H:%M:%S")
        data_formatada = data_convertida.strftime("%d/%m/%Y")

        linha["order_approved_at"] = data_formatada
        datas_aprovacao_formatadas += 1


#---------------------------------------------------------------
# RELATÓRIO DA FORMATAÇÃO DAS DATAS
#---------------------------------------------------------------

print("Total de datas de aprovação vazias:", datas_de_aprovacao_vazias)
print("Total de datas de aprovação formatadas:", datas_aprovacao_formatadas)
print("Exemplo de data de aprovação formatada:", pedidos[0]["order_approved_at"])
```

```python
#---------------------------------------------------------------
# RELATÓRIO FINAL DO PROJETO
#---------------------------------------------------------------

print("\n" + "=" * 60)
print("RELATÓRIO FINAL - PIPELINE DE LIMPEZA DE DADOS OLIST")
print("=" * 60)

#---------------------------------------------------------------
# RELATÓRIO DE PRODUTOS
#---------------------------------------------------------------

print("\n--- PRODUTOS ---")
print("Total de produtos processados:", len(produtos))
print("Quantidade de produtos sem categoria corrigidos:", len(cont_sem_categoria))

#---------------------------------------------------------------
# RELATÓRIO DE DIMENSÕES FÍSICAS
#---------------------------------------------------------------

print("\n--- DIMENSÕES FÍSICAS ---")
print("As colunas dimensionais tratadas foram:")
print("- product_length_cm")
print("- product_height_cm")
print("- product_width_cm")
print("- product_weight_g")
print("Regra aplicada: valores vazios foram substituídos pela média da própria coluna.")

#---------------------------------------------------------------
# RELATÓRIO DE PEDIDOS
#---------------------------------------------------------------

print("\n--- PEDIDOS ---")
print("Total de pedidos processados:", total_pedidos_processados)
print("Total de pedidos cancelados:", total_pedidos_cancelados)
print("Total de pedidos sem data de entrega:", total_pedidos_sem_data)
print("Pedidos sem data de entrega e status 'canceled':", total_pedidos_sem_data_status_cancelado)
print("Pedidos sem data de entrega e status diferente de 'canceled':", total_pedidos_sem_data_status_diferente)

#---------------------------------------------------------------
# VALIDAÇÃO DA HIPÓTESE DE NEGÓCIO
#---------------------------------------------------------------

print("\n--- VALIDAÇÃO DA REGRA DE NEGÓCIO ---")

if total_pedidos_sem_data_status_diferente == 0:
    print("Resultado: todas as datas de entrega vazias pertencem a pedidos cancelados.")
else:
    print("Resultado: existem pedidos sem data de entrega com status diferente de 'canceled'.")
    print("Portanto, a hipótese de que toda data vazia pertence a pedido cancelado não é totalmente verdadeira.")

#---------------------------------------------------------------
# RELATÓRIO DE DATAS DE APROVAÇÃO
#---------------------------------------------------------------

print("\n--- DATAS DE APROVAÇÃO ---")
print("Total de datas de aprovação vazias:", datas_de_aprovacao_vazias)
print("Total de datas de aprovação formatadas:", datas_aprovacao_formatadas)

#---------------------------------------------------------------
# EXEMPLOS DE EXCEÇÕES
#---------------------------------------------------------------

print("\n--- EXEMPLOS DE PEDIDOS SEM DATA E STATUS DIFERENTE DE CANCELED ---")

if len(exemplos_datas_vazias_outro_status) == 0:
    print("Nenhum exemplo encontrado.")
else:
    for exemplo in exemplos_datas_vazias_outro_status:
        print("ID do pedido:", exemplo["order_id"])
        print("Status:", exemplo["order_status"])
        print("Data de entrega:", exemplo["order_delivered_customer_date"])
        print("-" * 40)

#---------------------------------------------------------------
# FINALIZAÇÃO
#---------------------------------------------------------------

print("\n" + "=" * 60)
print("PROCESSAMENTO FINALIZADO COM SUCESSO")
print("=" * 60)
```
